##Usage per cell/site

###Preparing the data

```
rty = oqwe.drop_duplicates('sid')
er = 0

#Checking the similarity of the columns
for r, row in rty.iterrows():
  if (row['cid'] == row['sid']):
    er += 1

print (er)
print (len(rty))
#prints the percentage of the silimar against the none
print ((er/len(rty))*100)
```

With the above it can be concluded that CELL_ID is not similar to SITE_CODE thus only the data from the merged dataframe can be used for this analysis. Precisly 33.35% of CELL_ID and SITE_ID records are similar.

```
#Drops the records from the cell_geo dataset that could not corespond to and data on the telecom dataset 
zxc = zxc.dropna(subset=['p','v','ca'], axis=0)

comb = []
#Combining CELL_ID and SITE_ID  and makes them a combined object for unique distinction
for w , row in zxc.iterrows():
  comb.append(str(row['sid']) + "  " + str(row['cid']))

zxc['sidcid'] = comb
zxc
```

###Creating a frequency table using above data for analysis and analysis

####Collection of the data

```
#Creating and filling frequency table and populating

lis = zxc['sidcid'].unique()
idfre = pd.DataFrame(index = lis , columns = ['sidcid','voice', 'sms', 'data', 'total'])
idfre = idfre.fillna(0) 

for t, row in zxc.iterrows():
  idfre.loc[row['sidcid'],'sidcid'] = row['sidcid']
  if row['p'] == 'Voice':
    idfre.loc[row['sidcid'],'voice'] = idfre.loc[row['sidcid'],'voice'] + 1
    idfre.loc[row['sidcid'],'total'] = idfre.loc[row['sidcid'],'total'] + 1
  if row['p'] == 'sms':
    idfre.loc[row['sidcid'],'sms'] = idfre.loc[row['sidcid'],'sms'] + 1
    idfre.loc[row['sidcid'],'total'] = idfre.loc[row['sidcid'],'total'] + 1
  if row['p'] == 'Data':
    idfre.loc[row['sidcid'],'data'] = idfre.loc[row['sidcid'],'data'] + 1
    idfre.loc[row['sidcid'],'total'] = idfre.loc[row['sidcid'],'total'] + 1

idfre
```

```
#Splitting the combined column

k = idfre['sidcid'].apply(lambda x: pd.Series([i for i in x.lower().split('  ')]))
k.columns = ['SITE_ID', 'CELL_ID']

freqtable = pd.merge(idfre , k , left_index=True, right_index=True)
freqtable = freqtable.drop(columns = ['sidcid'])
freqtable.reset_index(drop = True , inplace = True)
freqtable
```

####Analysing

#####By frequency

```
#Which cell ID had the most usage by frequency
freqtable.sort_values('total', ascending =False)
```

```
#Which cell ID had the most usage by frequency for voice
freqtable.sort_values('voice', ascending =False)
```

```
#Which cell ID had the most usage by frequency by sms
freqtable.sort_values('sms', ascending =False)
```
